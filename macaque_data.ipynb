# Helper functions

```python
# Compute error
m = 5
n_avg = 1000

def compute_err(manifolds,m):
    err_all = np.zeros((len(manifolds),len(manifolds)))
    err_std = np.zeros((len(manifolds),len(manifolds)))
    for a in tqdm(range(len(manifolds))):
        Xa = manifolds[a]
        for b in range(len(manifolds)):
            Xb = manifolds[b]

            errs = []
            for _ in range(n_avg):
                perma = np.random.permutation(len(Xa))
                permb = np.random.permutation(len(Xb))

                xa,ya = np.split(Xa[perma],(m,))
                xb,yb = np.split(Xb[permb],(m,))
                w = (xa-xb).mean(0)
                mu = (xa+xb).mean(0)/2

                h = ya@w - w@mu
                err = (h<0).mean()
                errs.append(err)
            err_all[a,b] = np.mean(errs)
            err_std[a,b] = np.std(errs)
    np.fill_diagonal(err_all,np.nan)
    
    return err_all, err_std




def geometry(centers,Rs,Us,m):
    K = len(centers)
    P = Rs.shape[1]
    dists = np.sqrt(((centers[:,None] - centers[None])**2).sum(-1))
    dist_norm = dists / np.sqrt((Rs**2).sum(-1)[:,None] / P)

    Ds = np.sum(Rs**2,axis=-1)**2 / np.sum(Rs**4, axis=-1)

    # Center-subspace
    csa = []
    csb = []
    for a in range(K):
        for b in range(K):
            if a!=b:
                dx0 = centers[a] - centers[b]
                dx0hat = dx0 / np.linalg.norm(dx0)
                costheta_a = Us[a]@dx0hat
                csa.append((costheta_a**2 * Rs[a]**2).sum() / (Rs[a]**2).sum())
                costheta_b = Us[b]@dx0hat
                csb.append((costheta_b**2 * Rs[b]**2).sum() / (Rs[a]**2).sum())
            else:
                csa.append(np.nan)
                csb.append(np.nan)
    csa = np.stack(csa).reshape(K,K)
    csb = np.stack(csb).reshape(K,K)

    # Subspace-subspace
    ss = []
    for a in range(K):
        for b in range(K):
            if a!=b:
                cosphi = Us[a]@Us[b].T
                ss_overlap = (cosphi**2*Rs[a][:,None]**2*Rs[b]**2).sum() / (Rs[a]**2).sum()**2
                ss.append(ss_overlap)
            else:
                ss.append(np.nan)
    ss = np.stack(ss).reshape(K,K)

    css = (csa + csb/m) * dist_norm**2

    bias = ((Rs**2).sum(-1) / (Rs**2).sum(-1)[:,None] - 1)/m
    SNR = 1/2*(dist_norm**2 + bias/m)/ np.sqrt(1/Ds[:,None]/m + css + ss/m)
    
    return dist_norm, Ds, csa, csb, ss, SNR, bias
```

```python
# Compute error
m = 1
n_avg = 1000

def compute_err(manifolds,m):
    err_all = np.zeros((len(manifolds),len(manifolds)))
    err_std = np.zeros((len(manifolds),len(manifolds)))
    for a in tqdm(range(len(manifolds)),leave=False):
        Xa = manifolds[a]
        for b in range(len(manifolds)):
            Xb = manifolds[b]

            errs = []
            for _ in range(n_avg):
                perma = np.random.permutation(len(Xa))
                permb = np.random.permutation(len(Xb))

                xa,ya = np.split(Xa[perma],(m,))
                xb,yb = np.split(Xb[permb],(m,))
                w = (xa-xb).mean(0)
                mu = (xa+xb).mean(0)/2

                h = ya@w - w@mu
                err = (h<0).mean()
                errs.append(err)
            err_all[a,b] = np.mean(errs)
            err_std[a,b] = np.std(errs)
    np.fill_diagonal(err_all,np.nan)
    
    return err_all, err_std
```

```python
from scipy import stats

def plot_CI(x,y,c,fmt,alpha=1):
    plt.plot(x,y.mean(-1), fmt, c=c,alpha=alpha)
    CImin,CImax = stats.t.interval(0.95, len(y),
                     loc=y.mean(-1), scale=stats.sem(y,axis=-1))
    plt.fill_between(x,CImin,CImax,color=c, alpha=0.1)
```

```python
def scatter_CI(x,y,c,fmt,label,ax):
    CI = stats.t.interval(0.95, len(y),
                     loc=y.mean(-1), scale=stats.sem(y,axis=-1))
    yerr = np.abs(y.mean() - np.stack(CI)[:,None])
    plt.errorbar(x,y.mean(),yerr=yerr,color=c,marker=fmt,capsize=5,zorder=999)
    ax.annotate(label,(x+0.3,y.mean()), color=c,fontsize=16)
```

```python
def mshot_err(X):
    ''' Performs an m-shot learning experiment on a pair of shuffled manifolds X=(Xa,Xb).
    Args: 
        m: # training examples
        X: X=(Xa,Xb), a pair of (P,N) object manifolds, pre-shuffled along the zeroth axis.
    
    Returns:
        erra: m-shot learning error evaluated on manifold a
        errb: m-shot learning error evaluated on manifold b
    '''
    Xa,Xb = X
    xatrain, xatest = np.split(Xa, (m,))
    xa = xatrain.mean(0)
    xbtrain, xbtest = np.split(Xb, (m,))
    xb = xbtrain.mean(0)
    x = np.vstack([xa,xb])

    distsa = ((x[:,None] - xatest[None])**2).sum(-1)
    ha = distsa[1] - distsa[0]
#     ya = distsa.argmin(0)

    distsb = ((x[:,None] - xbtest[None])**2).sum(-1)
    hb = distsb[0] - distsb[1]
#     yb = distsb.argmin(0)

#     erra = (ya!=0)
#     errb = (yb!=1)

    return ha, hb
```

```python
n_avg = 1

K = len(manifolds)

errs_a = []
errs_b = []
errs_a_resnet = []
errs_b_resnet = []
errs_a_vgg = []
errs_b_vgg = []
errs_a_rn152 = []
errs_b_rn152 = []

for a in tqdm(range(K)):
    Xa = np.array(manifolds[a])
    Xaresnet = np.array(manifolds_resnet[a])
    Xavgg = np.array(manifolds_vgg[a])
    Xarn152 = np.array(manifolds_rn152[a])
    for b in range(a+1,K):
        Xb = np.array(manifolds[b])
        Xbresnet = np.array(manifolds_resnet[b])
        Xbvgg = np.array(manifolds_vgg[b])
        Xbrn152 = np.array(manifolds_rn152[b])

        erra = []
        errb = []
        erra_resnet = []
        errb_resnet = []
        erra_vgg = []
        errb_vgg = []
        erra_rn152 = []
        errb_rn152 = []

        for _ in range(n_avg):
            idxsa = np.random.permutation(len(Xa))
            idxsb = np.random.permutation(len(Xa))
            erratmp,errbtmp = mshot_err([Xa[idxsa],Xb[idxsb]])
            erratmp_resnet,errbtmp_resnet = mshot_err([Xaresnet[idxsa],Xbresnet[idxsb]])
            erratmp_vgg,errbtmp_vgg = mshot_err([Xavgg[idxsa],Xbvgg[idxsb]])
            erratmp_rn152,errbtmp_rn152 = mshot_err([Xarn152[idxsa],Xbrn152[idxsb]])
            erra.append(np.stack(erratmp))
            errb.append(np.stack(errbtmp))
            erra_resnet.append(np.stack(erratmp_resnet))
            errb_resnet.append(np.stack(errbtmp_resnet))
            erra_vgg.append(np.stack(erratmp_vgg))
            errb_vgg.append(np.stack(errbtmp_vgg))
            erra_rn152.append(np.stack(erratmp_rn152))
            errb_rn152.append(np.stack(errbtmp_rn152))
        errs_a.append(np.stack(erra))
        errs_b.append(np.stack(errb))
        errs_a_resnet.append(np.stack(erra_resnet))
        errs_b_resnet.append(np.stack(errb_resnet))
        errs_a_vgg.append(np.stack(erra_vgg))
        errs_b_vgg.append(np.stack(errb_vgg))
        errs_a_rn152.append(np.stack(erra_rn152))
        errs_b_rn152.append(np.stack(errb_rn152))
```
